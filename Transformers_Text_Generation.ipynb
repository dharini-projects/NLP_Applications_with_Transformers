```python
from transformers import pipeline
text_generator = pipeline("text-generation")
```

```python
prompt = "I am using Github to"
text_generator(prompt, max_length=30, num_return_sequences=5)
```

```python
prompt = "I can use GPT2 for"
text_generator(prompt, max_length=30, num_return_sequences=5)
```
